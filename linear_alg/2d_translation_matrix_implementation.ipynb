```python
import numpy as np
def translate_object(points, tx, ty):
    translated_points = np.asarray(points, dtype=np.float32)
    translated_points[:,0] += tx
    translated_points[:,1] += ty
    return translated_points
```

```python
points = [[0, 0], [1, 0], [0.5, 1]]
tx, ty = 2, 3

print(translate_object(points, tx, ty))
```
